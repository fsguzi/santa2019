```python
import pandas as pd
import numpy as np
import collections
```

```python
df = pd.read_csv('family_data.csv')
n = df.iloc[:,-1].sum()
```

```python
cost = []

def acc(cur,pre):
    cur += 125
    pre += 125
    return (cur-125)/400*pow(cur,0.5+abs(cur-pre)/50)

for i in range(176):
    cost.append([])
    for j in range(176):
        cost[-1].append(acc(i,j))
cost = np.array(cost)
```

```python
bk = np.ones(shape=(n+1,176))*10e9

for i in range(176):
    bk[n-125-i,i] = cost[i][i]

for i in range(99,0,-1):
    new = np.ones(shape=(n+1,176))*10e9
    for remain in range(i*125,n-(100-i)*125+1):
        best = (bk[remain]+cost).min(axis=1)
        most = min(300,remain)
        x = np.arange(remain-125,remain-most-1,-1)
        y = np.arange(most-125+1)
        new[x,y] = np.minimum(new[x,y],best[:most-125+1])
    bk = new
    print(i)
```
